```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
top_scores_df = pd.read_excel("top_20.xlsx")
top_scores_df.head(20)
```

```python
top_scores_df = top_scores_df.rename(columns={"Unnamed: 0": "uid", "score":"level"})
top_scores_df
```

```python
counts_per_level = top_scores_df.groupby(['level']).size()
counts_per_level.head(60)
```

```python
counts_per_level.plot.line()
plt.xlim([0,60])
```

```python
counts_per_level.plot.bar()
plt.xlim([0,35])
```

```python
top_scores_df["lastPlay"]=pd.to_datetime(top_scores_df['lastPlay'],unit='s')
top_scores_df
```

```python
level_46th = top_scores_df.loc[top_scores_df["level"] == 46]
level_46th
```

```python
level_upto45 = top_scores_df.loc[top_scores_df["level"] <= 45]
level_upto45
```

```python
total_level_attempts = level_upto45.groupby(["level"])["levelAttempts"].sum()
total_level_attempts
```

```python
level_attempts_median = level_upto45.groupby(["level"])["levelAttempts"].median()
level_attempts_median
```

```python
level_attempts_count = level_upto45.groupby(["level"])["uid"].count()
level_attempts_count
```

```python
level_merged = pd.merge(total_level_attempts, level_attempts_count, on='level')
level_merged
```
